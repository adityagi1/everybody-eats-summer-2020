# Get Zip codes for census tracts in the data

```python
# Create full tract codes for each census tract
pm_data['Tract_id'] = pm_data.Tract.str.extract('(\d*\.\d+|\d+)').astype(float)*100
pm_data['Tract_id'] = pm_data['Tract_id'].round(decimals=2).astype(int).astype(str).str.zfill(6)
pm_data['Tract_id'] = '17031' + pm_data['Tract_id'] 
```

```python
tract_to_zip.TRACT = tract_to_zip.TRACT.astype(str)
tract_to_zip.ZIP = tract_to_zip.ZIP.astype(int).astype(str)

pm_data_temp = pm_data.merge(tract_to_zip, left_on='Tract_id',
                                  right_on='TRACT', how='left')
```

```python
# Check tracts with no Zip codes mapped
pm_data_temp.loc[pm_data_temp.ZIP.isnull(), ['Tract','Tract_id','ZIP']]
```

```python
pm_data_zip = pm_data_temp.drop(columns=['Tract_id','TRACT','RES_RATIO','BUS_RATIO','OTH_RATIO'])
pm_data_zip.head()
```

# Calculated weighted average values for each Zip code

```python
pm_data_zip.columns
```

```python
# Check and impute missing values 
columns_na = pm_data_zip.columns[pm_data_zip.isna().any()].tolist()
print(columns_na)
columns_na.remove('ZIP')
columns_na.remove('TOT_RATIO')
print(columns_na)
```

```python
pm_data_zip[columns_na].apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
# Define the function for calculating weighted mean
def wavg(group, avg_name, weight_name):
    """ http://stackoverflow.com/questions/10951341/pandas-dataframe-aggregate-function-using-multiple-columns
    In rare instance, we may not have weights, so just return the mean. Customize this if your business case
    should return otherwise.
    """
    d = group[avg_name]
    w = group[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return d.mean()
```

```python
# Calculate weighted averages for all the variables
avg_household_size = pm_data_zip.groupby("ZIP").apply(wavg, "avg_household_size", "TOT_RATIO")
avg_travel_time_to_work = pm_data_zip.groupby("ZIP").apply(wavg, "avg_travel_time_to_work", "TOT_RATIO")
computer_access = pm_data_zip.groupby("ZIP").apply(wavg, "computer_access", "TOT_RATIO")
life_expectancy = pm_data_zip.groupby("ZIP").apply(wavg, "life_expectancy", "TOT_RATIO")
median_age = pm_data_zip.groupby("ZIP").apply(wavg, "median_age", "TOT_RATIO")
median_home_loan_amount = pm_data_zip.groupby("ZIP").apply(wavg, "median_home_loan_amount", "TOT_RATIO")
median_home_value = pm_data_zip.groupby("ZIP").apply(wavg, "median_home_value", "TOT_RATIO")
median_household_income = pm_data_zip.groupby("ZIP").apply(wavg, "median_household_income", "TOT_RATIO")
median_leverage_ratio = pm_data_zip.groupby("ZIP").apply(wavg, "median_leverage_ratio", "TOT_RATIO")
num_jobs = pm_data_zip.groupby("ZIP").apply(wavg, "num_jobs", "TOT_RATIO")
num_housing_units = pm_data_zip.groupby("ZIP").apply(wavg, "num_housing_units", "TOT_RATIO")
proportion_bachelors_degree = pm_data_zip.groupby("ZIP").apply(wavg, "proportion_bachelors_degree", "TOT_RATIO")
proportion_disabled = pm_data_zip.groupby("ZIP").apply(wavg, "proportion_disabled", "TOT_RATIO")
proportion_hispanic = pm_data_zip.groupby("ZIP").apply(wavg, "proportion_hispanic", "TOT_RATIO")
prop_men = pm_data_zip.groupby("ZIP").apply(wavg, "prop_men", "TOT_RATIO")
prop_nonwhite = pm_data_zip.groupby("ZIP").apply(wavg, "prop_nonwhite", "TOT_RATIO")
prop_nonenglish_speaking = pm_data_zip.groupby("ZIP").apply(wavg, "prop_nonenglish_speaking", "TOT_RATIO")
prop_families_poverty = pm_data_zip.groupby("ZIP").apply(wavg, "prop_families_poverty", "TOT_RATIO")
prop_students_in_school = pm_data_zip.groupby("ZIP").apply(wavg, "prop_students_in_school", "TOT_RATIO")
social_vulnerability_index = pm_data_zip.groupby("ZIP").apply(wavg, "social_vulnerability_index", "TOT_RATIO")
thiel_racial_segregation_index = pm_data_zip.groupby("ZIP").apply(wavg, "thiel_racial_segregation_index", "TOT_RATIO")
```

```python
summary = pd.DataFrame(data=dict(s1=avg_household_size, 
                                        s2=avg_travel_time_to_work,
                                        s3=computer_access,
                                        s4=life_expectancy,
                                        s5=median_age, 
                                        s6=median_home_loan_amount,
                                        s7=median_home_value,
                                        s8=median_household_income,
                                        s9=median_leverage_ratio, 
                                        s10=num_jobs,
                                        s11=num_housing_units,
                                        s12=proportion_bachelors_degree,
                                        s13=proportion_disabled, 
                                        s14=proportion_hispanic,
                                        s15=prop_men,
                                        s16=prop_nonwhite,
                                        s17=prop_nonenglish_speaking,
                                        s18=prop_families_poverty,
                                        s19=prop_students_in_school,
                                        s20=social_vulnerability_index,
                                        s21=thiel_racial_segregation_index))
summary.columns = ['avg_household_size',
       'avg_travel_time_to_work', 'computer_access', 'life_expectancy',
       'median_age', 'median_home_loan_amount', 'median_home_value',
       'median_household_income', 'median_leverage_ratio', 'num_jobs',
       'num_housing_units', 'proportion_bachelors_degree',
       'proportion_disabled', 'proportion_hispanic', 'prop_men',
       'prop_nonwhite', 'prop_nonenglish_speaking', 'prop_families_poverty',
       'prop_students_in_school', 'social_vulnerability_index',
       'thiel_racial_segregation_index']

summary.head()
```
