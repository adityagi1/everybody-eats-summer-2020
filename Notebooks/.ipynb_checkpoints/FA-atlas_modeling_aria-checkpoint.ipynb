# Main takeaway:
- This notebook build 3 linear regression models to predict the food insecurity # in 2020 in each census tract in Evanston, using the data about low income and low food access population in 2015.
- Variables that have high and positive correlations with the food insecurity # in 2020 are: 'the census tract total population in 2015', 'whether the census tract is a low income and low access tract in 2015(measured at 1/2 mile for urban areas and 10 miles for rural areas)', 'whether the census tract is a low income tract in 2015', 'tract poverty rate in 2015', 'the number of low-income population in the tract in 2015', 'food insecurity rate in 2018' and 'food insecurity number in 2018.'
- The chosen model (with in-sample R-squared score = 0.77)to have the following predictors: 'the census tract total population in 2015', 'whether the census tract is a low income and low access tract in 2015(measured at 1/2 mile for urban areas and 10 miles for rural areas)', 'whether the census tract is a low income tract in 2015', 'tract poverty rate in 2015', 'the number of low-income population in the tract in 2015.'
     + With other variables remaining constant:
         * as the census tract total population in 2015 increases by 1 unit, the 2020 food insecurity number tends to increase by 0.1;
         * the census tracts that are low income and low access tracts in 2015 (measured at 1/2 mile for urban areas and 10 miles for rural areas) tend to have 96 more food insecurity people in 2020 compared with other census tracts;
         * the census tracts that are low income tracts in 2015 tend to have 58 more food insecurity people in 2020 compared with other census tracts;
         * as the tract poverty rate in 2015 increases by 1%, the 2020 food insecurity number tends to increase by 86;
         * as the number of of low-income population in the tract in 2015 increases by 1 unit, the 2020 food insecurity number tends to increase by 0.1.

# Import packages and load the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
feeding_america_data = pd.read_excel('./Food Insecurity Rates by Census tracts.xlsx')
feeding_america_data.head(5)
```

```python
# Transfer the tract information in feeding america data to census tract number
feeding_america_data['Tract_number'] = \
feeding_america_data.Tract.str.extract('(\d*\.\d+|\d+)').astype('float')*100
feeding_america_data['Tract_number'] = \
feeding_america_data.Tract_number.astype(int).astype(str)

# Correct the variable types
feeding_america_data['[C] 2020 Food Insecurity #'] = \
feeding_america_data['[C] 2020 Food Insecurity #'].str.replace(",", "").astype(int)
feeding_america_data['2018 Food Insecurity #'] = \
feeding_america_data['2018 Food Insecurity #'].str.replace(",", "").astype(float)

feeding_america_data.head(5)
```

```python
atlas_data = pd.read_csv('./atlas/USDA_atlas.csv')
atlas_data.head(5)
```

```python
# Transfer the tract information in atlas data to census tract number
atlas_data[['census_tract_id']] = atlas_data['census_tract_id'].astype(int).astype(str)
atlas_data['Tract_number'] = atlas_data.census_tract_id.str[5:]

atlas_data.head()
```

# Merge data and calculate the correlation matrix

```python
food_insecurity_atlas = atlas_data.merge(feeding_america_data, on='Tract_number', how='left')
```

# Build models with potentially influential variables

```python
# Find out the variables with the highest correlations with the food insecurity number in 2020
corr_df = data_corr.corr()
corr_df1 = corr_df[['[C] 2020 Food Insecurity #']][abs(corr_df['[C] 2020 Food Insecurity #'])>=0.6]
corr_df1
```

```python
potential_features = corr_df1.index.tolist()
print(potential_features)
```

```python
data_corr = food_insecurity_atlas[potential_features]
data_corr.corr()
```

```python
fig, ax = plt.subplots(figsize=(10, 8)) 
sns.heatmap(data_corr.corr(), ax= ax)
```

## Model 1

```python
features = ['Population, tract total', 
  'Low income and low access tract measured at 1/2 mile for urban areas and 10 miles for rural areas', 
  'Low income tract', 
  'Tract poverty rate', 
  'Tract low-income population, number', 
  '2018 Food Insecurity', 
  '2018 Food Insecurity #']
X = food_insecurity_atlas[features]
y = food_insecurity_atlas[['[C] 2020 Food Insecurity #']]

regressor1 = LinearRegression()  
regressor1.fit(X.values, y.values)

r_sq = regressor1.score(X.values, y.values)
print('The R-squared score is: {}'.format(r_sq))
print('\n')

coeff_df = pd.DataFrame(data=regressor1.coef_, columns=X.columns).transpose().rename(columns={0: 'coefficient'})
print('The correlation matrix is:')
print(coeff_df)
print('\n')

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
print('The VIF scores are:')
print(vif)
```

- The VIF scores above show that we should remove '2018 Food Insecurity #' from the model.
